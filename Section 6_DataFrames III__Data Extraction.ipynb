## Rename index label or columns in dataframe
 - Two primary ways to use rename
     - you can use the mapper and axis parameter
         - We specified which axis we want to target, eithere the row axis or the index label or column axis, which means the column headers right here to map
         - Mapper we can pass a dictionary. In that dictionary, the keys are going to represent the existing labels in that axis and the values for those dictionary.
     - or you can use index and columns parameter independently

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.rename(mapper = {"GoldenEye": "Golden Eye",
                      "The World Is Not Enough": " Best Vond Movie Ever"})
```

```python
bond.rename(mapper = {"GoldenEye": "Golden Eye",
                      "The World Is Not Enough": " Best Vond Movie Ever"})

bond.rename(mapper = {"GoldenEye": "Golden Eye",
                      "The World Is Not Enough": " Best Vond Movie Ever"}, axis = 0)

bond.rename(mapper = {"GoldenEye": "Golden Eye",
                      "The World Is Not Enough": " Best Vond Movie Ever"}, axis = 'rows')

bond.rename(mapper = {"GoldenEye": "Golden Eye",
                      "The World Is Not Enough": " Best Vond Movie Ever"}, axis = 'index')  

bond.rename(index = {"GoldenEye": "Golden Eye",
                      "The World Is Not Enough": " Best Vond Movie Ever"})

# All of the five is the same!
# If we wanbt to change this permantely, we will need to add inplace = True
```

```python
bond.head(1)
```

```python
bond.rename(mapper = {"Year": "Release Date", "Box Office":"Revenue"}, axis =1)
bond.rename(mapper = {"Year": "Release Date", "Box Office":"Revenue"}, axis ='columns')
bond.rename(columns = {"Year": "Release Date", "Box Office":"Revenue"}, inplace = True)
```

## A review of the .apply() method on single columns

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + "MILLIONS!"

bond['Box Office'] = bond['Box Office'].apply(convert_to_string_and_add_millions)
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

## The .apply() method with row values

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == 'Roger Moore' and budget >40:
        return "Enjoyable"
    else:
        return 'I have no clue'

bond.apply(good_movie, axis = "columns") # Think it traversing the columns in the same row
```

## The .copy() method
- WIll prevent the small subset that we extract will not impact the original data

```python
bond = pd.read_csv('pandas/jamesbond.csv', index_col = 'Film')
bond.sort_index(inplace = True)
bond.head(3)
```

```python
Directors = bond['Director']
Directors.head(3)
```

```python
Directors["A View to a Kill"] = "Mister John Glen"
```

```python
Directors.head(3) # The value has been changed!
```

```python
bond.head(3) # The original dataframe got affected as well!
```

```python
directors = bond['Director'].copy() # This will be a brand new series
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mister John Glen"
```

```python
directors.head(3)
```
